```python
dic1=manager.dict()
dic2=manager.dict()
dic3=manager.dict()

f = open('/content/words.txt', 'r')
arr=f.readlines()
l=[]
for i in arr:
    l.append(i.rstrip("\n"))
print(l[:23])
print(len(l))
l=l[:10000]
len(l)
```

```python
def clue1(w1, w2):
    x = [value for value in w1 if value in w2] # character wise comparison
    return len(x)
```

```python
def clue2(w1):
    w2=["a","e","i","o","u"]
    x = [value for value in w1 if value in w2]
    return len(x)    # len(x) will contain how many vowels are contained in w1.
```

```python
def c1p1(dic1,start,end=-1):
    for i in l[start:end]:
        temp=0
        for j in l:
            temp+=clue1(i,j)    # which means temp will contain the total similarity of ith word with respect to every other word in the list.
        dic1.update({i:temp})
```

```python
def c2(dic2,l):
    for i in l:
        temp=0
        temp+=clue2(i)
        dic2.update({i:temp})   # so temp will contain number of vowels contained in ith word.
```

```python
def c3(dic3,l):
    d3 = sorted(l, key=len, reverse = True)
    h=1
    for i in d3:
        dic3.update({i:h*0.33})
        h+=1
```

```python
len(l)
```

```python
# Serial Execution
start = time()

# Write your code here
c1p1(dic1,0,len(l))
c2(dic2,l)
c3(dic3,l)

print(time()-start)
```

```python
# Parallel Execution

# Create two process that executes clp1 on half lists.
t=len(l)
result1=Process(target=c1p1, args=(dic1,0,int(t/2)))
result2=Process(target=c1p1, args=(dic1,int(t/4),int(t/2)))



# Create one process runs c2 that operates in complete list l.
result3=Process(target=c2, args=(dic2,l))

# Create one process runs c3 that operates in complete list l.
result4=Process(target=c3, args=(dic3,l))

start=time()
result1.start()
result2.start()
result3.start()
result4.start()

result1.join()
result2.join()
result3.join()
result4.join()

print(time()-start)
```
